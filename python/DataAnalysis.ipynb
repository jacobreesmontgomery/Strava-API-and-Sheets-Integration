# STEP ONE: DATA ACQUISITION AND FORMATTING

```python
# Importing the necessary libraries
import pandas as pd
import datetime as dt
```

```python
# Reading in the CSV file and displaying the first five activities
activities_df = pd.read_csv("C:/Users/17178/Desktop/GITHUB_PROJECTS/Strava-API-and-Sheets-Integration/python/data/GoonsActivities.csv", header=0)
activities_df.head(5)
```

```python
# Viewing data types of each column
activities_df.dtypes
```

```python
# Converting full date
activities_df["FULL DATE"] = activities_df["FULL DATE"].astype("string")
activities_df["FULL DATE"] = pd.to_datetime(activities_df["FULL DATE"], yearfirst=False, dayfirst=False)

# Converting time
activities_df["TIME"] = pd.to_datetime(activities_df["TIME"]).dt.strftime('%H:%M %p')
# Annoyingly, when I do strftime('%H:%M %p'), things sort correctly below but I don't get the format I want
# BUT, when I do strftime('%I:%M %p'), I get the format I want, but then things don't sort correctly :/
# Desire: Format I want AND it sorts correct. 

# Sorting to be safe
activities_df = activities_df.sort_values(by=["FULL DATE", "TIME"], ascending=[False, False]) # Extra precaution (things should be sorted properly as is)
```
